### Project
This project seeks to explore the price and price spread of the cryptoasset/cryptocurrency Bitcoin.

### Data
Using Bitcoin price data (2020 YTD, 63 days) downloaded from coinmarketcap.com (https://coinmarketcap.com/currencies/bitcoin/historical-data/?start=20200101&end=20200304), I created a new variable called "spread" using the existing "high" and "low" prices of the cryptoasset for that day by calculating the difference between the two to obtain the price spread for the day. I also created another new variable called "price change" using the existing "close" and "open" prices of the cryptoasset for that day by calculating the difference between the two.

Since this is comming directly from a website specializing in cryptoassets, I do not anticipate any data quality issues and believe that this dataset would be perfect for my analysis.

```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

data = pd.read_csv('Bitcoin.csv')
data.head()
```

```python
data.info()
```

The variables are in their desired Dtypes so no conversion is needed.

```python
data2 = data[['Open', 'High', 'Low', 'Close']]
data2['Spread'] = data2['High'] - data2['Low']
data2['Price Change'] = data2['Close'] - data2['Open']
data2.head()
```

```python
data2.describe()
```

```python
%matplotlib notebook
import seaborn as sns

sns.lineplot(err_style=None, data=data2)
```

As we can see from the visualization above, the open, high, low, and close of the cryptoasset are all seemingly close together. This is supported by the "spread" and "price change" variables, both of which are close to the bottom of the visualization. The fact that they are so close to the bottom of the visualization tells us that the price spread and price change of the cryptoasset are relatively low to to it's open, high, low, and close prices. Although the price spread is relatively stable day to day (with the exception of a few peaks and drops) the price change fluctuates a lot, most likely a result of current economic conditions and the stock market.
